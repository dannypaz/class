```python
print("Percentage of readmit_30: {:.2%}".format((hospital_df.readmit_30.value_counts()[1]/hospital_df.readmit_30.value_counts().sum())))
```

```python
print("Cost of readmit_30 fines: ${:0,.2f}".format(hospital_df.readmit_30.value_counts()[1] * 8000))
```

### By using CareTracker, we would avoid 40% of `readmit_30`:

```python
print("Number of patients avoiding readmit_30: {:0,.0f}".format(hospital_df.readmit_30.value_counts()[1] * 0.4))
```

```python
no_avoid = hospital_df.readmit_30.value_counts()[1] * 0.6
```

```python
print("Cost of CareTracker: ${:0,.2f}".format(hospital_df.readmit_30.value_counts().sum() * 1200))
```

```python
print("Cost of readmit_30 fines with CareTracker: ${:0,.2f}".format(no_avoid * 8000))
```

#### Total cost difference by using CareTracker:

##### before:

```python
print("Cost without CareTracker: ${:0,.2f}".format(hospital_df.readmit_30.value_counts()[1] * 8000))
```

#### after:

```python
print("Cost with CareTracker: ${:0,.2f}".format((hospital_df.readmit_30.value_counts().sum() * 1200 + no_avoid * 8000)))
```

##### difference:

```python
print("Cost difference: ${:0,.2f}".format(hospital_df.readmit_30.value_counts()[1] * 8000 - (hospital_df.readmit_30.value_counts().sum() * 1200 + no_avoid * 8000)))
```

### So, providing CareTracker to everybody and with a 40% reduction of `readmit_30`, overall costs will increase by $2,064,800.
#### We need to so something else to reduce costs.
