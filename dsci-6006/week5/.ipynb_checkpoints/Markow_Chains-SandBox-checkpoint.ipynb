```python
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('purchases.txt', sep="\t", header=None, names=['customer_id', 'purchase_amount', 'date_purchase'])
```

```python
df.head()
```

```python
df['date_purchase_formatted'] = df['date_purchase'].apply(lambda x: dt.datetime.strptime(x,"%Y-%m-%d"))
```

```python
df.info()
```

```python
df.head()
```

```python
df.shape
```

### We create a cross-tab table:

```python
pd.crosstab(df['customer_id'], df['date_purchase_formatted'], margins=True)
```

```python
df['date_purchase'] =  pd.to_datetime(df['date_purchase'])
```

```python
df['year_purchase'] = df['date_purchase'].apply(lambda x: x.year)
```

```python
df['month_purchase'] = df['date_purchase'].apply(lambda x: x.month)
```

```python
df.head()
```

```python
import datetime as dt

# http://www.marketingdistillery.com/2014/11/02/rfm-customer-segmentation-in-r-pandas-and-apache-spark/

NOW = dt.datetime(2016,1,1)

rfmTable = df.groupby('customer_id').agg({ 'date_purchase_formatted': lambda x: (NOW - x.max()).days, # R
                                            'purchase_amount': lambda x: x.sum()}) # M
```

```python
plt.figure(figsize=(14, 7))
purchase_plot = df.groupby('date_purchase').sum()['purchase_amount'].plot(rot=90)
purchase_plot.set(xlabel='Date', ylabel='Purchase Amount')
plot_mean = df.groupby('date_purchase').sum()['purchase_amount'].mean()
plot_std = df.groupby('date_purchase').sum()['purchase_amount'].std()
purchase_plot.axhline(y=plot_mean, color='black')
purchase_plot.axhline(y=plot_mean+plot_std*1.5, color='red', ls='dashed')
purchase_plot.axhline(y=plot_mean-plot_std*1.5, color='red', ls='dashed')
purchase_plot.plot();
```

```python
plt.figure(figsize=(14, 7))
customer_plot = df.groupby('customer_id').sum()['purchase_amount'].plot(rot=90)
customer_plot.set(xlabel='Customer ID', ylabel='Purchase Amount')
plot_mean = df.groupby('customer_id').sum()['purchase_amount'].mean()
plot_std = df.groupby('customer_id').sum()['purchase_amount'].std()
customer_plot.axhline(y=plot_mean, color='black')
customer_plot.axhline(y=plot_mean+plot_std*1.5, color='red', ls='dashed')
customer_plot.axhline(y=plot_mean-plot_std*1.5, color='red', ls='dashed')
customer_plot.plot();
```
