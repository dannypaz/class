### Discrete and Continuous Sample Spaces

* A sample space is ** finite ** if it has a finite number of elements; otherwise, it is ** infinite **


* A sample space is ** discrete ** if the elements in it can be listed, e.g. $1, \ 2, \ 3, \dots$  
    In mathematical language, a sample space is discrete if it is ** countable **
    
    
* A sample space is ** continuous ** if there are no gaps between the elements, so the elements cannot be listed, e.g. the interval $[0, \ 1]$

Combining Events
------

Formulating events in terms of sets gives us the power of set theory to describe all possible ways of combining events. 

- Union
- Intersection
- Complement

Union
-----

<center><img src="images/union.png" height="500"/></center>

The union of two events A and B, denoted A ∪ B, is the set of outcomes that belong either to A, to B, or to both. 

In words, A ∪ B means "A and/or B." 

Example: 6 on either die #1 or die #2 (or both)

Intersection
-----

<center><img src="images/intersection.png" height="500"/></center>

The intersection of two events A and B, denoted A ∩ B, is the set of outcomes that belong __both__ to A and to B.

In words, A ∩ B means “A and B.”

Example: 6 on __both__ dice

Complement
-----

<center><img src="images/com.png" height="500"/></center>

The complement of an event B is the set of outcomes that do not belong to A. 

In words, $A^c$ means "not A."

Example: 6 on die #2 **but not** on die #1

A little more on Complement
-----

$P(A) + P(A^c) = 1$

It is a useful tool because it is often easier to find the probability that an event did __not__ occur

Independent Events
----

Events A and B are __independent__ if and only if:  

$$P(A \cap B) = P(A)P(B)$$

Dependent Events
----

If events A and B are __dependent__: 

P(A and B) = P(A ∩ B) = P(A)P(B|A) = P(B)P(A|B)

\* You can always assume events are dependent. If they are _independent_, you are just doing a little extra work.

Conditional! 


A conditional probability is denoted by P(A|B).   
This phrase is read: _The probability that A will occur given that B is known to have occurred._

Pancake Puzzle
-----

<center><img src="http://i3.mirror.co.uk/incoming/article697628.ece/ALTERNATES/s615/Susan-Boyle-flipping-pancakes-at-St-Augustines-High-School-in-Edinburgh.jpg" height="500"/></center>

Say I’m making pancakes. The first one is burned on both sides. The second one is better—only one side burned. The third one is okay on both sides. I choose a pancake at random and observe that one side is not burned. What are the chances the other side isn’t burned, either?

The chances are two out of three. 

The pancake you chose was one of these two: 

1. The okay/burned one 
2. The okay/okay one. 

Given this choice, you could be looking at any of three “okay” sides. 

Of those three sides, two of them are okay on the other side, too.

Venn Diagrams Limitations
-----

Venn diagrams are generally useful for up to 3 events, although they are not used to provide formal proofs. 

For more than 3 events, the diagram might not be able to represent all possible overlaps of events.

When would you simulate probabilities?
------

Sometimes it is inconvenient to explicitly define a sample space.

- The sample space is very large and complicated
- The sample space is infinite
- We feel more confident in writing a program to simulate the situation, 
rather than one to enumerate the complete sample space.

    - Intersections
    - Complements
    - Addition
    - Multiplication

- P(A and B) = P(A)· P(B|A)
- Do not be intimidated (or fooled) by statistical challenges. Systematically define and solve.

------
